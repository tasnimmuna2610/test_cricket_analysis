### Renaming columns

```python
df = df.rename(columns={'Mat':'Matches', 
                        'Inns':'Innings',
                        'Wkts': 'Wickets',
                        'Ave': 'Average',
                        'Econ': 'Economy'})

display(df.head())
```

### Removing columns

```python
df.drop('BBI', axis=1, inplace=True)

display(df.head())
```

### Removing BBM Column


```python
df.drop('BBM', axis=1, inplace=True)

display(df.head())
```

### Finding how many players played for ICC

```python
# For this we will need to seperate the country column from the player
df = df.rename(columns={'SR': 'Strike_Rate'})
# splitting the 'Player' column to get the information about 'Country'
df_player = df['Player'].str.split("(", expand=True)

# concatinating 'Country' with the main dataframe
df = pd.concat([df, df_player], axis=1)

# dropping the 'Player' columns
df = df.drop('Player', axis=1)

# renaming the column names
df = df.rename(columns={0: 'Player',
                        1: 'Country'})

# remove the ")" from the 'Country' column
df['Country'] = df['Country'].str.replace(")", "")

# rearrange the columns
new_col_sequence = ['Player', 'Span', 'Matches', 'Innings', 'Runs', 'Wickets',
                    'Average', 'Economy', 'Strike_Rate', 5, 'Country']
df = df[new_col_sequence]

display(df.head())
```

```python
# defining the function with condition
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Country'].apply(icc_check)
```

```python
df['played_for_ICC'].value_counts()
```

### Finding the player who had played for the longest period of time

```python
# splitting the 'Span' column based on the "-"
df_span = df['Span'].str.split("-", expand=True)

# conccatinating the new dataframe with the main dataframe
df = pd.concat([df, df_span], axis=1)

# renaming the newly created column names
df = df.rename(columns={0: "start_year",
                        1: "end_year"})

# removing the "Span" column
df = df.drop("Span", axis=1)

display(df.head())
```

```python
#checking the datatype
df.info()
```

```python
#converting data type
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')

print(df.info())
```

```python
# Making a column for years the player played
df['years_played'] = df['end_year'] - df['start_year']

df = df.drop(['start_year', "end_year"], axis=1)

display(df.head(10))
```

### Top 5 batsmen: Highest number of years played

```python
df.sort_values(by="years_played", ascending = False).head(5)
```

### Top 5 batsmen: Lowest number of years played

```python
df.sort_values(by="years_played", ascending = True).head(5)
```

### Number of Australian Bowlers present in this dataset

```python
# defining the function with condition
def aus__player_check(x):
    if "AUS" in x:
        return "Australian"
    else:
        return "Not Australian"
```

```python
df['australian_bowler'] = df['Country'].apply(aus__player_check)
```

```python
df['australian_bowler'].value_counts()
```

### Finding for Bangladeshi player in the dataset

```python
# defining the function with condition
def bengali__player_check(x):
    if "BAN" in x:
        return "Bengali Players Found"
    else:
        return "Not found"
```

```python
df['bengali_player'] = df['Country'].apply(bengali__player_check)
```

```python
df['bengali_player'].value_counts()
```
